## What a Neural Network is and how it works

A Neural Network is a computational model inspired by the structure and functioning of the human brain. It consists of interconnected nodes, or "neurons," organized into layers. Information travels through the network, undergoing weighted transformations and passing through activation functions at each neuron.

During training, the network learns to **adjust these weights**, optimizing its ability to map inputs to desired outputs.

The utilities of neural networks pass through regression and classification models, image and speech pattern recognition, emotion analysis, language translation.

### Neural Network Components

image.png


**Neuron**: These are the basic units of a neural network. Each neuron performs a mathematical operation on the inputs and produces an output.

**Layer**: Neurons are organized into layers. In a typical neural network, there are three types of layers:
 - Input Layer: Receives input signals and passes the information to hidden layers.
 - Hidden Layers: These layers perform calculations and transform inputs into learned features. Networks with multiple hidden layers are called deep neural networks.
 - Output Layer: Produces network outputs after processing information through the hidden layers.

**Inputs**: Independent features of the training dataset. Each input is introduced into a neuron of the input layer.

**Outputs**: They are divided into two types:
  - Expected output: The true value for each instance of the dataset
  - Predicted output: The prediction of the NN. This output is compared with the expected one to measure the quality of the prediction.

**Weights**: Each connection between neurons has a weight, which is the **main trainable parameter of the NN**.

**Bias**: Each neuron has a bias that works as the intercept in Regression. It is also a trainable parameter.

**Activation function**: Activation functions introduce nonlinearities into the network. Common functions include ReLU (Rectified Linear Unit), Sigmoid, and Hyperbolic Tangent, softmax.

**Loss function**: Also known as cost function, measures how far the model's predictions are from the actual labels. Compares the two types of outputs. The network adjusts its weights to minimize this function during training.


In short, during training, information passes through the entire network, from the input layer to the output layer. Inside the NN occurs multiple simple mathematical operation and transofrmations that map the patterns found in the input data. When the transformed data arrives to the output layer, the loss function measures the quality of the prediction.

In a **dense** NN every single neuron in a layer is connected with every neuron in the next layer (and previous). This connections are weighted.

Neurons in the input layer receive the raw information from the dataset and spread them into every neuron of the first hidden layer. Neurons in the hidden layer have as input the outputs of the neurons of the previous layer. Each input, $x_i$ , is multiplied (weighted), $w_i$ , and added. Then the bias ($b$) is added. Next, the result is passed through the activation function, which returns the output of the neuron (input for the neurons of the next layer).

$$\sum_{i=1}^{n} (x_i * w_i) + b --> Activation function --> Neuron-output$$





## More Terminology

**Epoch**: Each time the training data is passed through the entire network it is called Epoch. Usually, when training a NN, multiple epochs are parametrized (10, 50, 100, 200...)

**Batch size**: To optimize the calculus efficiency and model convergence, input data is passed in lots. Each epoch is divided into several batches.

**Optimizer**: Algorithm who fit the weights during training process. It minimize the cost function. Some exemples are: Gradient Descent, Stochastic Gradient Descent and Adam.

**Learning rate**: It controls the weight fit speed. How much weights are changed per iteration.

**momentum**: In the context of optimization algorithms, momentum is a technique used to accelerate the training of neural networks and other machine learning models. It helps the model to continue moving in the same direction even when the gradients are small, speeding up convergence and improving the chances of finding the optimal solution.

**Regularization**: As in regression regularization penalizes the cost function to avoid overfitting.

**Dropout**: Tunable parameter. Some neurons are randomly selected to not been modified during an epoch. Each epoch the dropped out neurons change. It allows avoid overfitting.

**Early Stopping**: Is a regularization technique that stops training when the model’s performance on a validation set stops improving, helping to prevent overfitting. 



## Backpropagation

Backpropagation, short for **backward propagation of errors**, is a fundamental algorithm used to train artificial neural networks. It's a supervised learning algorithm that trains the neural network by minimizing the error between its predictions and the actual target values. Here's how backpropagation works step by step:

1.-Initialization:

Initialize the neural network with random weights and biases. These weights and biases are adjusted during training to minimize prediction errors.

2.-Forward Pass:

 - Feed a training example (input) forward through the network. Calculate the predicted output.
 - For each neuron in each layer, calculate the weighted sum of inputs, apply an activation function to this sum, and pass the result to the next layer as output.
 - Calculate the error between the predicted output and the actual target values using a loss function (e.g., mean squared error for regression problems or cross-entropy loss for classification problems).

3.-Backward Pass (Backpropagation):

 - Compute the gradient of the loss function with respect to the network's weights and biases. This gradient represents how much the loss would increase or decrease if a particular weight or bias is adjusted slightly.
 - Use the **chain rule** of calculus to calculate these gradients layer by layer, starting from the output layer and moving backward through the network. This step is why it's called "backpropagation."
 - Update the weights and biases of the network using an optimization algorithm (commonly gradient descent). The update rule typically involves subtracting a fraction of the gradient from the current weights and biases. **Learning rate** determines the step size of these updates.
 - Repeat the forward and backward passes for multiple epochs (iterations over the entire dataset) until the network's predictions become sufficiently accurate or the loss converges to a minimum.

4.-Validation and Testing:

 - After training, the model's performance is evaluated on a separate  test dataset (to estimate its generalization to unseen data).
 - If the model performs well on the test dataset, it can be deployed for making predictions on new, unseen data.

The backpropagation algorithm enables neural networks to learn from data by iteratively adjusting their weights and biases to minimize prediction errors. It's a foundational technique in deep learning and underpins the training process of most modern neural network architectures.

## Vanishing and exploding gradient problems

The vanishing gradient problem and the exploding gradient problem are two issues that can occur during the training of deep neural networks, particularly in networks with many layers.

1.-Vanishing Gradient Problem:

- **Issue**: In deep networks, during backpropagation, gradients can become exceedingly small as they are propagated backward through the layers. This can happen due to the multiplication of many small gradient values (in the range [0, 1]) during the chain rule. When gradients become too small, they essentially vanish, meaning the weights of the earlier layers are not updated effectively, and those layers stop learning.
- **Consequence**: Layers early in the network don't learn meaningful features, leading to poor overall network performance, especially in deep architectures.
- **Solution**: Techniques like ReLU activation functions, careful weight initialization methods, and normalization techniques like **batch normalization** can mitigate the vanishing gradient problem by maintaining gradients at a reasonable scale during training.

2.-Exploding Gradient Problem:

- **Issue**: Conversely, gradients can also explode during backpropagation, especially in deep networks. This occurs when the gradients become very large, often due to unstable weight initializations or very high learning rates. Large gradients lead to large updates to the weights, causing the model to overshoot the optimal solution and fail to converge.
- **Consequence**: The model's weights are adjusted too drastically, causing the optimization process to become unstable, and the model fails to converge to a solution.
- **Solution**: **Gradient clipping** is a common technique used to address exploding gradients. It involves rescaling gradients if they exceed a certain threshold, preventing them from becoming excessively large. Additionally, careful **weight initialization** and choosing **appropriate learning rates** are essential to avoid this problem.

  - `He` initialization for ReLU y `Xavier/Glorot` for tanh/sigmoid

Both vanishing and exploding gradients highlight the challenges of training deep networks. Modern architectures and training techniques, including proper weight initialization, activation functions like ReLU, batch normalization, and gradient clipping, have been developed to mitigate these issues, enabling the successful training of deep neural networks.




### Batch normalization

Batch normalization is a technique used in deep learning to **improve training stability and performance**. It works by **normalizing the outputs** of a layer over a mini-batch (before getting in the activation function), so that each neuron's activations have a mean close to zero and a standard deviation close to one. This helps reduce internal covariate shift and **mitigates issues like vanishing or exploding gradients** during training.

Batch normalization does not directly prevent weights from becoming too small or too large. Instead, by keeping activations in a stable range, it supports more effective gradient propagation and helps optimization converge faster and more reliably.

By ensuring that the activations are within a consistent range, the optimization algorithm can adapt more easily, as it does not need to account for large fluctuations in input scales across layers.

Functions like the sigmoid or tanh have derivatives close to zero when their inputs are very large or very small.

By normalizing activations around zero, they are kept in regions where the derivative is not so small, helping prevent the gradients from vanishing.

## Optimizer algorithms


Certainly, here's a closer look at the first four optimization algorithms:

**1.-Gradient Descent (GD)**:

 - Description: Gradient Descent calculates the derivative of the loss function with respect to the model's parameters and moves in the opposite direction of the gradient to minimize the loss.
 - Advantages: Simplicity and intuitive concept.
 - Disadvantages: May converge slowly, especially on complex and high-dimensional spaces.

**2.-Stochastic Gradient Descent (SGD):**

 - Description: SGD updates the model's parameters after each training example, making it faster and more suitable for large datasets.
 - Advantages: Faster convergence, suitable for online learning and large datasets.
 - Disadvantages: Noisy updates, might oscillate around the minimum.

**3.-Mini-Batch Gradient Descent:**

 - Description: Mini-Batch Gradient Descent combines the concepts of GD and SGD. It updates parameters using small random subsets (mini-batches) of the training data.
 - Advantages: Balance between the efficiency of SGD and stability of GD.
 - Disadvantages: Requires tuning batch size.

**4.-Adam (Adaptive Moment Estimation):**

 - Description: Adam uses the concept of momentum. Momentum helps accelerate the convergence process. It updates the weights by mini-batch. Adam adapts the learning rates of each parameter individually based on their historical gradients and squared gradients.
 
     An "accumulated velocity" is calculated:

    $
    v_t = \beta \cdot v_{t-1} + (1 - \beta) \cdot \nabla \mathcal{L}(w)
    $

    The weight is updated using that velocity:

    $
    w \leftarrow w - \eta \cdot v_t
    $

    **Where:**

    - $v_t$: accumulated velocity (momentum)  
    - $\nabla \mathcal{L}(w)$: current gradient  
    - $\eta$: learning rate  
    - $\beta$: momentum factor (typically 0.9)

     - Advantages: Adaptive learning rates for each parameter, combines momentum for faster convergence, works well for noisy or sparse gradients.
     - Disadvantages: Requires tuning hyperparameters.

  In the previous optimizers like SGD or Mini-batch the weights are updated using this formula:

    $w_1 \leftarrow w_1 - \eta \cdot \frac{\partial \mathcal{L}}{\partial w_1}$


Each algorithm has its strengths, and the choice depends on factors like the dataset size, the noise level in the gradients, and computational resources. Empirical testing often determines which algorithm performs best for a specific task.

**Using python** the difference between GD, SGD and minibatch is parametrized with `batch_size = n`

## Activation functions

ReLU (Rectified Linear Unit), tanh (hyperbolic tangent), and sigmoid are activation functions commonly used in neural networks. Here's a brief overview of each:

**1.-ReLU (Rectified Linear Unit)**:

 - Range: $[0, inf)$

 - Function: $f(x) = max(0,x)$
 
 <img src="./../assets/ReLU.png" alt="Second Image" style="width: 25%;"/>

ReLU is the most widely used activation function in deep learning. It replaces all negative values with zero and leaves positive values unchanged. It **mitigates the vanishing gradient problem**, allowing for faster convergence during training. It's **computationally efficient** and often leads to better performance in deep networks.

ReLU outputs zero for all negative inputs, which means many neurons are inactive (output = 0) at any given time.
This introduces sparsity in the network, making it more efficient and sometimes **helping with generalization** because it avoids relying on every possible detail of the training data.



**2.-Hyperbolic Tangent**:

 - Range: $(-1,1)$


 - Function:  $f(x) = \tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$
 
 <img src="./../assets/tanh.png" alt="Second Image" style="width: 20%;"/>


tanh squashes the output to be between -1 and 1. It is zero-centered, making it easier to model inputs that have strongly negative, zero, and positive values. 

By helping the next layer’s neurons receiving inputs with zero mean, often improves the convergence speed.

However, it is affected by the **vanishing gradients** and it is **computationally expensive**.

**3.-Sigmoid**:

Range: $(0,1)$


Function: $f(x) = \frac{1}{1 + e^{-x}}$

**4.-Softmax**: It is commonly used in the **output layer of neural networks** for **multi-class classification** tasks. It transforms a vector of raw scores (*logits*) into a **probability distribution**, where each value lies between 0 and 1, and the sum of all values equals 1. # Loss Functions

###  Mean Squared Error (MSE)

**Definition:**

$
\text{MSE} = \frac{1}{N} \sum_{i=1}^N (y_i - \hat{y}_i)^2
$

- $y_i$: true value  
- $\hat{y}_i$: predicted value  
- $N$: number of samples


- Regression problems (predicting continuous values).  
- Simple and smooth.  
- Penalizes large errors more (quadratically).


### Mean Absolute Error (MAE)

#### Definition:

$
\text{MAE} = \frac{1}{N} \sum_{i=1}^N |y_i - \hat{y}_i|
$

- $y_i$: true value  
- $\hat{y}_i$: predicted value  
- $N$: number of samples


- Regression tasks where robustness to outliers is important. Less sensitive to outliers than MSE.


###  Binary Cross-Entropy (BCE)

**Definition:**

For each sample with label $y \in \{0,1\}$ and prediction $\hat{y}$:

$
\text{BCE} = - \left( y \log(\hat{y}) + (1 - y) \log(1 - \hat{y}) \right)
$

**Use case:**  
- Binary classification problems.  
- Output layer usually has **1 neuron + sigmoid** activation.



###  Categorical Cross-Entropy (CCE)

**Definition:**

For multi-class classification with true class index $k$ and predicted probabilities $\hat{y}_j$:

$
\text{CCE} = - \sum_{j} y_j \log(\hat{y}_j)
$

where $y_j$ is 1 for the correct class and 0 otherwise (one-hot encoding).

- Multi-class classification.  
- Output layer usually has **softmax** activation producing proability distribution.
- Generalizes binary cross-entropy to multiple classes.  
- **Requires one-hot encoded labels in the target variable**.
  - For example, for 3 classes, class 2 would be:

    $y = [0, 1, 0]$

    The predicted output $\hat{y}$ is a probability distribution over all classes (from softmax), e.g.,

    $\hat{y} = [0.2, 0.7, 0.1]$

    CCE then calculates the loss focusing on the predicted probability of the correct class.



## Trade-off Small/Large batch size

The choice between small and large mini-batches in deep learning is not a one-size-fits-all decision and depends on various factors. Here are some considerations:

**Advantages of Small Mini-Batches**:

 - Better Generalization: Small mini-batches introduce noise, which can act as a regularizer, potentially improving the model's ability to generalize to unseen data.
 - Escape Local Minima: Noise in small mini-batches can help the optimization process escape sharp local minima, leading to potentially better solutions.
 - Improved Exploration: Small mini-batches allow the optimization algorithm to explore the loss landscape more thoroughly, potentially finding better solutions.

**Advantages of Large Mini-Batches**:

 - Faster Training: Larger mini-batches take better advantage of parallel processing capabilities, leading to faster training times, especially on modern GPUs.
 - Stable Gradients: Larger mini-batches provide more stable gradient estimates, which can result in a smoother optimization process and faster convergence.

**Considerations**:

 - Noise vs. Stability: Small mini-batches introduce noise, which can help with generalization, but too much noise might hinder convergence.
 - Memory and Computational Resources: Large mini-batches require more memory. If memory is limited, using excessively large mini-batches could lead to memory errors.

**Best Practice**:

 - It's often best to empirically test different mini-batch sizes to find the one that balances the benefits of regularization and exploration (with small mini-batches) against the stability and computational efficiency (with large mini-batches). This testing should be done on a validation dataset to ensure the model's generalization to unseen data is optimal.

In summary, the choice between small and large mini-batches is problem-dependent. It's essential to strike a balance based on the specific problem, the available computational resources, and empirical testing to ensure the best performance of the trained model.

### Classification - Diabetes

```python
diabetes = pd.read_csv('../../../Data/Deep Learning/Deep Learning/pima-indians-diabetes.csv', delimiter=';')
diabetes.head(3)
```

```python
diab_X_train, diab_X_test, diab_y_train, diab_y_test = train_test_split(diabetes[['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI',
                                              'DiabetesPedigreeFunction','Age']],
                                                    diabetes['Outcome'], train_size = 0.8, random_state = 0)
diab_X_train.head(3)
```

### Regression - Car Miles Galon

```python
cars = pd.read_csv('../../../Data//Deep Learning/Deep Learning/car_miles_per_galon.csv', sep = ';')
cars.tail(3)
```

```python
cars.shape
```

```python
cars_origin = pd.get_dummies(cars['origin'], dtype = int)
cars_origin.columns = ['origin1','origin2','origin3']
cars = cars.merge(cars_origin, right_index = True, left_index = True, how='left')
```

```python
cars_X_train, cars_X_test, cars_y_train, cars_y_test = train_test_split(cars[['cylinders','displacement','horsepower',
                                                               'weight','acceleration','model year','origin1','origin2','origin3']],
                                                    cars['mpg'], train_size = 0.8, random_state = 0)
cars_X_train.head(3)
```

```python
cars_y_train.shape
```

```python
column_to_keep_train = cars_X_train[['origin1','origin2','origin3']]
cars_X_train = cars_X_train.drop(columns=['origin1','origin2','origin3'])
original_index_train = cars_X_train.index

column_to_keep_test = cars_X_test[['origin1','origin2','origin3']]
cars_X_test = cars_X_test.drop(columns=['origin1','origin2','origin3'])
original_index_test = cars_X_test.index

scaler = StandardScaler()
car_scaler = scaler.fit(cars_X_train)
train_car_scaler = car_scaler.transform(cars_X_train)
test_car_scaler = car_scaler.transform(cars_X_test)

# Crear un nuevo DataFrame con las características estandarizadas
X_train_car_scaler = pd.DataFrame(train_car_scaler, columns=cars_X_train.columns, index = original_index_train)
X_test_car_scaler = pd.DataFrame(test_car_scaler, columns=cars_X_test.columns, index = original_index_test)

# Concatenar la columna no estandarizada de nuevo al DataFrame
X_train_car_scaler = pd.concat([X_train_car_scaler, column_to_keep_train], axis=1)
X_test_car_scaler = pd.concat([X_test_car_scaler, column_to_keep_test], axis=1)
X_test_car_scaler.head(3)
```

## MLP for Classification

```python
MLP_Clas = Sequential()
# MLP_Clas.add(BatchNormalization())
MLP_Clas.add(Dense(10, input_shape=(8,), kernel_initializer='he_uniform', activation='sigmoid'))
MLP_Clas.add(BatchNormalization())

MLP_Clas.add(Dense(16, activation='sigmoid', kernel_regularizer=l2(0.001)))
MLP_Clas.add(BatchNormalization())
MLP_Clas.add(Dropout(0.2))

MLP_Clas.add(Dense(10, activation='sigmoid', kernel_regularizer=l1(0.001)))
MLP_Clas.add(BatchNormalization())

MLP_Clas.add(Dense(1, activation='sigmoid'))

# Learning rate and optimizer
learning_rate = 0.1
optimizer = Adam(learning_rate=learning_rate) # SGD(learning_rate=learning_rate)

early_stop = EarlyStopping(
    monitor='val_loss',      # métrica a observar
    patience=15,              # número de épocas que esperamos mejora
    restore_best_weights=True,  # vuelve a los pesos con mejor val_loss,
    min_delta = 0.001
)

# compile the keras model
MLP_Clas.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy', Recall()]) # metrics=['accuracy']

history_clas = MLP_Clas.fit(diab_X_train_sc, diab_y_train, validation_data=(diab_X_test_sc, diab_y_test), 
                            epochs=90, batch_size=32, callbacks = early_stop) # , callbacks = early_stop
# evaluate the keras model
_, accuracy_tr, recall_tr = MLP_Clas.evaluate(diab_X_train_sc, diab_y_train)
_, accuracy_ts, recall_ts = MLP_Clas.evaluate(diab_X_test_sc, diab_y_test)
print('Accuracy train: %.2f; Recall train: %.2f' % (accuracy_tr*100, recall_tr*100))
print('Accuracy test: %.2f; Recall test: %.2f' % (accuracy_ts*100, recall_ts*100))
```

```python
MLP_Clas.summary()
```

```python
pyplot.subplot(211)
pyplot.title('Accuracy')
pyplot.plot(history_clas.history['accuracy'], label='train')
pyplot.plot(history_clas.history['val_accuracy'], label='test')
pyplot.legend()
```

```python
pyplot.subplot(211)
pyplot.title('Loss')
pyplot.plot(history_clas.history['loss'], label='train')
pyplot.plot(history_clas.history['val_loss'], label='test')
pyplot.legend()
```

```python
predictions = MLP_Clas.predict(diab_X_test_sc, verbose = 0).round(0)
conf_mat = confusion_matrix(diab_y_test, predictions)
print(tabulate(conf_mat,headers = ['pred Diabetes No','pred Diabetes Yes'], showindex = ['real Diabetes No','real Diabetes Yes'],
               tablefmt = 'fancy_grid'))

print(classification_report(diab_y_test, predictions))
```

# Backpropagation Example — Simple Neural Network

## Network Architecture

- **Input**: 1 neuron → $x \in \mathbb{R} $
- **Hidden Layer**: 2 neurons with ReLU activation
- **Output Layer**: 1 neuron with sigmoid activation
- **Loss Function**: Binary Crossentropy

---

## Forward Pass

### Hidden Layer (2 neurons):

$
z_1 = w_1 x + b_1, \quad h_1 = \text{ReLU}(z_1)
$
$
z_2 = w_2 x + b_2, \quad h_2 = \text{ReLU}(z_2)
$

### Output Layer:

$
z_3 = w_3 h_1 + w_4 h_2 + b_3
$
$
\hat{y} = \sigma(z_3) = \frac{1}{1 + e^{-z_3}}
$

---

##  Loss Function (Binary Crossentropy):

$
\mathcal{L}(y, \hat{y}) = - \left[ y \log(\hat{y}) + (1 - y) \log(1 - \hat{y}) \right]
$

---

##  Backward Pass — Derivative with respect to $ w_1 $

We apply the **chain rule** step-by-step:

$
\frac{\partial \mathcal{L}}{\partial w_1} =
\frac{\partial \mathcal{L}}{\partial \hat{y}} \cdot
\frac{\partial \hat{y}}{\partial z_3} \cdot
\frac{\partial z_3}{\partial h_1} \cdot
\frac{\partial h_1}{\partial z_1} \cdot
\frac{\partial z_1}{\partial w_1}
$

---

##  Derivatives

1. **Binary crossentropy**:
$
\frac{\partial \mathcal{L}}{\partial \hat{y}} = \frac{\hat{y} - y}{\hat{y}(1 - \hat{y})}
$

2. **Sigmoid function**:
$
\frac{\partial \hat{y}}{\partial z_3} = \hat{y}(1 - \hat{y})
$

3. **Linear combination**:
$
\frac{\partial z_3}{\partial h_1} = w_3
$

4. **ReLU activation**:
$
\frac{\partial h_1}{\partial z_1} = \mathbb{1}_{z_1 > 0}
$

5. **Linear combination**:
$
\frac{\partial z_1}{\partial w_1} = x
$

---

##  Final Gradient

$
\frac{\partial \mathcal{L}}{\partial w_1} = 
(\hat{y} - y) \cdot w_3 \cdot \mathbb{1}_{z_1 > 0} \cdot x
$

This gives the gradient of the loss with respect to weight $w_1 $.

---

## Final weight update

$\eta$ is negative because we want to minime the error, not to maximize it. If the lost is negative means that the function on that point is descending, so if we go to the right (increasing $w$) the error will be lower. Analogously, if the lost is positive, the function on that $w$ point is ascending, in that case if we increase $w$ the error will increase even more, that's why we would want to decrease $w$.

$w_1 \leftarrow w_1 - \eta \cdot \frac{\partial \mathcal{L}}{\partial w_1}$
