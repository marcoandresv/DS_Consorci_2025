```python
import requests
import pandas as pd
```

```python
# Parámetros de la consulta
species_name = "Apis mellifera iberiensis"
start_year = 2015
end_year = 2025  # Año actual o superior
limit = 300  # Número de registros por consulta (máx 300)

# URL base de la API de GBIF
gbif_url = "https://api.gbif.org/v1/occurrence/search"

# Parámetros para la solicitud
params = {
    "scientificName": species_name,  # Nombre científico
    "year": f"{start_year},{end_year}",  # Rango de años
    "limit": limit,  # Número de registros
    "hasCoordinate": "true",  # Solo con coordenadas geográficas
}

# Realizar la solicitud GET a la API
response = requests.get(gbif_url, params=params)
```

```python
data = response.json()
occurrences = data.get("results", [])
df=pd.DataFrame(occurrences)
df.columns
```
