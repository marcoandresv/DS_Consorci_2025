## 1. Introducci√≥n al Aprendizaje No Supervisado

### 1.1 Visi√≥n General del Aprendizaje No Supervisado

El aprendizaje no supervisado es un tipo de aprendizaje autom√°tico que se ocupa del reconocimiento de patrones y el an√°lisis exploratorio de datos. A diferencia del aprendizaje supervisado, los algoritmos de aprendizaje no supervisado no reciben datos de salida etiquetados. En cambio, estos algoritmos exploran la estructura de los datos para extraer informaci√≥n significativa sin orientaci√≥n.

**Puntos Clave:**
- **Exploraci√≥n de Datos:** El aprendizaje no supervisado se utiliza a menudo para descubrir la estructura subyacente de los datos.
- **Agrupamiento:** Una tarea com√∫n de aprendizaje no supervisado es agrupar datos en cl√∫steres basados en m√©tricas de similitud o distancia.
- **Reducci√≥n de Dimensionalidad:** Reducir el n√∫mero de variables en consideraci√≥n, que puede dividirse en selecci√≥n de caracter√≠sticas y extracci√≥n de caracter√≠sticas.

**Aplicaciones en el Mundo Real:**
- **Segmentaci√≥n de Mercado:** Identificar grupos distintos de clientes para adaptar estrategias de marketing.
- **Gen√≥mica:** Clasificar genes con funcionalidades similares sin conocimiento previo.
- **Detecci√≥n de Anomal√≠as:** Identificar patrones inusuales que no se ajustan al comportamiento esperado.

### 1.2 Distinci√≥n entre Aprendizaje Supervisado y No Supervisado

![](https://www.researchgate.net/publication/351953193/figure/fig3/AS:11431281117150742@1675395484096/Supervised-and-unsupervised-machine-learning-a-Schematic-representation-of-an.png)

El aprendizaje supervisado y el no supervisado representan dos enfoques fundamentales del aprendizaje autom√°tico, cada uno con sus casos de uso y m√©todos √∫nicos.

**Aprendizaje Supervisado:**
- **Datos Etiquetados:** Los algoritmos de aprendizaje supervisado se entrenan con datos etiquetados, lo que significa que cada ejemplo de entrenamiento se empareja con una etiqueta de salida.
- **Retroalimentaci√≥n Directa:** El objetivo es aprender un mapeo de entradas a salidas, con retroalimentaci√≥n directa sobre la precisi√≥n de la predicci√≥n.
- **Tareas de Predicci√≥n:** Las tareas comunes incluyen la clasificaci√≥n y la regresi√≥n.

**Aprendizaje No Supervisado:**
- **Sin Etiquetas:** En contraste, el aprendizaje no supervisado trata con datos que no tienen etiquetas hist√≥ricas.
- **Sin Retroalimentaci√≥n:** El sistema no se le dice la respuesta 'correcta'; debe descubrir qu√© se le est√° mostrando.
- **Tareas de Descubrimiento:** El objetivo es explorar los datos y encontrar alguna estructura dentro.

**Ilustraci√≥n Comparativa:**
Imagina el aprendizaje supervisado como un estudiante aprendiendo con un profesor para resolver problemas matem√°ticos con una clave de respuestas conocida. Por el contrario, el aprendizaje no supervisado es como un detective tratando de encontrar patrones y conexiones en un caso sin pistas previas.

**Conectando Ambos:**
Aunque el aprendizaje supervisado y el no supervisado pueden parecer distintos, pueden ser complementarios. El aprendizaje semisupervisado, por ejemplo, utiliza tanto datos etiquetados como no etiquetados, lo cual es particularmente √∫til cuando adquirir etiquetas es costoso.

## 2. T√©cnicas de Agrupamiento

### 2.1 Estudio Profundo del Agrupamiento K-means

El agrupamiento K-means es una t√©cnica ampliamente utilizada en el aprendizaje autom√°tico no supervisado para dividir un conjunto de datos en grupos distintos, conocidos como 'clusters'. Es particularmente √∫til cuando tienes datos que se agrupan naturalmente pero no tienen etiquetas que gu√≠en el agrupamiento.

[VIDEO: Enlace K-means](https://www.youtube.com/watch?v=4b5d3muPQmA&ab_channel=StatQuestwithJoshStarmer)

#### Entendimiento del Agrupamiento K-means:

Piensa en K-means como una forma de organizar puntos de datos en grupos basados en su similitud. Aqu√≠ tienes una analog√≠a simple: imagina que tienes una mezcla de bolas de diferentes colores esparcidas por el suelo, y quieres organizarlas en grupos del mismo color. El agrupamiento K-means ser√≠a como encontrar el punto central para cada grupo de color y luego agrupar las bolas en funci√≥n de qu√© punto central est√°n m√°s cerca.

![](https://miro.medium.com/v2/resize:fit:1200/1*rw8IUza1dbffBhiA4i0GNQ.png)

### Pasos del Algoritmo Explicados:

1. **Inicializaci√≥n**: Se inicializan los centroides al azar. Tambi√©n se puede elegir los puntos exactos.
2. **Asignaci√≥n**: Cada punto de datos en tu conjunto 'decide' a qu√© centroide se agrupa.
3. **Actualizaci√≥n**: Cada centroide se reconfigura debido a la nueva asignaci√≥n de datos.
4. **Repetir**: Los pasos 2 y 3 se siguen repitiendo. Con cada repetici√≥n, los grupos se refinan m√°s hasta que no necesitan moverse m√°s.

#### Objetivo de K-means:

El objetivo es minimizar la distancia entre los puntos en un cluster y su centroide. Cuando los centroides dejan de moverse (o el movimiento es m√≠nimo), significa que los clusters son tan ajustados y precisos como pueden ser, basados en el algoritmo K-means.

### C√≥mo determinar el n√∫mero K de clusters √≥ptimo

##### M√©todo del Puntaje Silhouette


El Puntaje Silhouette es una medida utilizada para determinar el grado cohesi√≥n dentro de los clusters y la separaci√≥n entre diferentes clusters. El objetivo es calcular la medida silhouette para varios K y comparar los resultados.

Para cada punto \( i \), se calcula:

- a: distancia media entre el punto \( i \) y todos los dem√°s puntos en el mismo *cluster* (baja ‚áí buena cohesi√≥n).
- b: distancia m√≠nima media entre el punto \( i \) y todos los puntos en el *cluster* m√°s cercano distinto (alta ‚áí buena separaci√≥n).

El coeficiente de Silhouette para el punto \( i \) se define como:

$ s(i) = \frac{(b-a)}{max(a, b)}$

- $s_i \in [-1,1]$  
- si $ s_i \approx 1 $ el punto est√° bien agrupado.
- si $ s_i \approx 0 $ el punto est√° cerca del borde entre dos grupos.
- si $ s_i < 1 $: el punto puede estar mal clasificado.


#### M√©todo del codo (Elbow method)

El m√©todo Elbow calcula las distancias entre cada punto y su centroide. A esta distancia se le llama WCSS (Within Cluster Sum of Squares). Nuevamente debe calcularse esta m√©trica para diferentes escenarios con K clusters. Posteriormente representar en un gr√°fico donde el eje X representa el n√∫mero de clusters y el eje Y la WCSS. 

$ WCSS = \sum_{i=1}^{k} \sum_{x \in C_i} \| x - \mu_i \|^2$



- ùê∂ ùëñ: Conjunto de puntos en el cluster *i*.
- ùúá ùëñ: Centroide (media) del cluster *i*.
- ‚à• ùë• ‚àí ùúá ùëñ ‚à•¬≤: Distancia cuadrada entre el punto *x* y su centroide.






### Ejemplo de clusterizaci√≥n con K-means

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
from sklearn.metrics import silhouette_score

# 1. Cargar datos Iris
iris = load_iris()
X = iris.data

# 2. Inicializar listas
wcss = []
silhouette_scores = []
k_range = range(2, 11)  # silhouette no se define para k=1

# 3. Calcular WCSS e √≠ndice de Silhouette
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    wcss.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X, labels))

# 4. Graficar ambos en subplots o doble eje
fig, ax1 = plt.subplots(figsize=(8, 5))

color_wcss = 'tab:blue'
ax1.set_xlabel('N√∫mero de clusters (k)')
ax1.set_ylabel('WCSS (inercia)', color=color_wcss)
ax1.plot(k_range, wcss, 'o-', color=color_wcss, label='WCSS')
ax1.tick_params(axis='y', labelcolor=color_wcss)
ax1.set_xticks(list(k_range))
ax1.grid(True)

# Eje secundario para silhouette
ax2 = ax1.twinx()
color_sil = 'tab:orange'
ax2.set_ylabel('Coeficiente Silhouette', color=color_sil)
ax2.plot(k_range, silhouette_scores, 's--', color=color_sil, label='Silhouette')
ax2.tick_params(axis='y', labelcolor=color_sil)

# T√≠tulo y mostrar
plt.title('M√©todo del Codo + Silhouette (Iris Dataset)')
fig.tight_layout()
plt.show()

```

```python
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
kmeans.fit(X)

# 3. Predecir etiquetas para cada punto
labels = kmeans.predict(X)

# 4. Visualizaci√≥n de los resultados
plt.figure(figsize=(8, 6))

# Gr√°fico en 2D usando las dos primeras caracter√≠sticas
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o', edgecolor='black')

# Centroides de los clusters
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, marker='X', label='Centroides')

# A√±adir t√≠tulo y etiquetas
plt.title('Clustering KMeans con K=3 (Iris Dataset)')
plt.xlabel('Longitud del s√©palo')
plt.ylabel('Ancho del s√©palo')
plt.legend()
plt.grid(True)
plt.show()
```

### Interpretaci√≥n

 - Los puntos est√°n coloreados seg√∫n los clusters asignados por el modelo K-Means. Cada color representa un cluster distinto.

 - Los X rojos son los centroides de los tres clusters calculados por el modelo. Estos puntos son los que minimizan la suma de distancias dentro de su grupo.
 
 - En el gr√°fico de dispersi√≥n se est√°n mostrando los puntos con respecto a 2 variables (Ancho y longitud del s√©palo), a pesar de no ser las √∫nicas variables utilizadas para la definici√≥n de los conglomerados. Por eso aparecen algunos puntos que parecen mal asignados

## M√©todo jer√°rquico de creaci√≥n de Conglomerados

El m√©todo de clustering jer√°rquico es una t√©cnica de agrupamiento que organiza los datos en una jerarqu√≠a de grupos o "clusters", sin necesidad de definir de antemano el n√∫mero de clusters.

### Pasos del Algoritmo Explicados:

 - Inicialmente cada punto es considerado un cluster.
 - Los dos clusters m√°s cercanos se unen entre s√≠
 - Se repite el punto anterior hasta que todos los puntos se han unido en uno solo

## Definici√≥n de m√°s cercano

Existe 4 m√©todos para determinar qu√© es m√°s o menos cercano. Todos ellos eval√∫an :

 - Single Linkage (enlace simple): Se calcula la distancia entre todos los puntos del cluster A y del cluster B, as√≠ como las distancias entre todos las posibles uniones entre clusters. De manera que, con el m√©todo Single, la distancia entre dos clusters es la **distancia m√≠nima** entre cualquier par de puntos, uno de cada cluster. Los conglomerados que muestren una distancia m√≠nima menor ser√°n los candidatos a unirse.
 
 - Complete Linkage (enlace completo): De igual manera que anteriormente, excepto que ahora √±a distancia entre dos clusters es la **distancia m√°xima** entre cualquier par de puntos, uno de cada cluster. Igualmente, los conglomerados que muestren una distancia m√°xima menor ser√°n los candidatos a unirse.
 
 - Average Linkage (enlace promedio): La distancia entre dos clusters es el promedio de todas las distancias entre pares de puntos entre ambos clusters. Aquellos que obtengan una distancia promedio menor se unir√°n en un solo conglomerado.
 
 - Ward: En lugar de usar distancias directas, mide el **aumento en la varianza total intra-cluster** si se fusionan dos clusters. Calcula todas las posibles combinaciones de uniones y sus aumentos de varianza. La uni√≥n que menos aumento suponga ser√° la candidata.
 
## Definiciones de distancias

 - Distancia Eucl√≠dea: La m√°s habitual. Mide la longitud del segmento recto entre dos puntos en un espacio n-dimensional.

 - Distancia Manhattan: Suma de las distancias absolutas por dimensi√≥n. Como caminar en una cuadr√≠cula.
 
 - Distancia de Mahalanobis: Mide la distancia considerando la correlaci√≥n entre variables.
 
 - Distancia del coseno: Mide la diferencia de direcci√≥n entre dos vectores, no su magnitud. √ötil cuando el √°ngulo entre vectores es m√°s importante que su distancia (por ejemplo, en texto o datos esparsos).

## 2.3 M√©todo DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

**DBSCAN** es un algoritmo de clustering basado en la densidad, lo que significa que forma clusters de puntos que est√°n **densamente conectados** entre s√≠, y puede **identificar puntos ruidosos** o **outliers** que no pertenecen a ning√∫n cluster. A diferencia de los m√©todos como K-means, DBSCAN no necesita que se especifique el n√∫mero de clusters de antemano. 

Con respecto a los otros dos algoritmos anteriores, es mejor, ya que puede encontrar **clusters de forma arbitraria**, lo que lo hace adecuado para datos con formas complejas (no necesariamente esf√©ricas), y detecta outliers (puntos ruidosos). Como contrapunto, requiere una correcta parametrizaci√≥n de dos par√°metros.

<img src="https://miro.medium.com/v2/resize:fit:1400/1*rfi9uHjGPdNgXgxe9xWvVw.png" width="50%"/>


### Par√°metros 
Para definir "puntos que est√°n densasamente conectados" se utilizan **dos par√°metros**

1. **eps (Œµ)**:  
   - El radio de b√∫squeda para encontrar vecinos cercanos. Es la **distancia m√°xima** entre dos puntos para que se consideren como parte del mismo cluster. Si la distancia entre dos puntos es menor que $\epsilon$, los puntos son vecinos. Cuanto menor es $\epsilon$ m√°s conglomerados se crear√°n y m√°s puntos ser√°n definidos como ruido (outlier). Por el contrario, cuanto m√°s grande menos cluster tendremos y con mayor varianza interna.
  
2. **min_samples**:  
   - El n√∫mero m√≠nimo de puntos requeridos para formar un cluster. Un punto se considera parte de un cluster si tiene al menos **min_samples** puntos dentro de su radio $\epsilon$. Igualmente, cuanto menor el n√∫mero de min_samples se generar√°n m√°s clusters m√°s peque√±os y m√°s outliers.
   
![](https://media.geeksforgeeks.org/wp-content/uploads/20190418023034/781ff66c-b380-4a78-af25-80507ed6ff26.jpeg)

## **Tipos de puntos en DBSCAN**
1. **Puntos n√∫cleo**:  
   - Un punto es un **punto n√∫cleo** si tiene al menos **min_samples** puntos dentro de su vecindad \( \epsilon \).

2. **Puntos frontera**:  
   - Un **punto frontera** tiene menos de **min_samples** puntos en su vecindad \( \epsilon \), pero est√° dentro del vecindario de un punto n√∫cleo.

3. **Puntos ruido**:  
   - Si un punto no es ni n√∫cleo ni frontera, se considera **punto ruido** o **outlier**.



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import DBSCAN

# Cargar el dataset Iris
iris = load_iris()
X = iris.data

# Aplicar DBSCAN
db = DBSCAN(eps=0.5, min_samples=5)
y_db = db.fit_predict(X)

# Visualizar los clusters encontrados
plt.scatter(X[:, 0], X[:, 1], c=y_db, cmap='viridis')
plt.title('DBSCAN Clustering')
plt.xlabel('Longitud del s√©palo')
plt.ylabel('Ancho del s√©palo')
plt.show()

# Ver la cantidad de outliers (-1 son los outliers)
print(f"Outliers: {np.sum(y_db == -1)}")
```
